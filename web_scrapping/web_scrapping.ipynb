```python
import selenium
from bs4 import BeautifulSoup as bs
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.options import Options
import time


driver = webdriver.Chrome()
driver.get("https://www.imdb.com/chart/top/?ref_=chtmvm_ql_3")
```

```python
movie_names=driver.find_elements(By.TAG_NAME,"h3")
other_data1=driver.find_elements(By.CLASS_NAME,"ipc-rating-star--rating")
other_data2=driver.find_elements(By.CLASS_NAME,"ipc-rating-star--voteCount")
other_data3=driver.find_elements(By.CLASS_NAME,"sc-b4f120f6-7")
```

```python
ratings=[]
for e in other_data1:
   ratings.append(e.text)
```

```python
movies=[]
for e in movie_names:
   movies.append(e.text)
driver.close()
```

```python
from bs4 import BeautifulSoup
with open ("page_data.html","r",encoding="utf-8")as f:
    file=f.read()
soup=BeautifulSoup(file,"lxml")

file=soup.prettify()
with open ("pretty_soup.txt","w",encoding="utf-8") as f:
    f.write(file)
```

```python
elements=soup.find_all(class_="sc-b4f120f6-7")
otherdata=[]
for e in elements:
    otherdata.append(e.text)

chunks = [otherdata[i:i+3] for i in range(0, len(otherdata), 3) if (i<=201)]
```

```python
import pandas as pd
df=pd.DataFrame()
df['Movies']=movies[0:68]
df['ratings']=ratings[0:68]
df['year']=[ chunk[0] for chunk in chunks]
df['hours']=[ chunk[1] for chunk in chunks]
df['content_rating']=[ chunk[2] for chunk in chunks]
```

```python
df=df.style.set_properties(**{'text-align': 'left'}).set_table_styles(
    [{'selector': 'th', 'props': [('text-align', 'left')]}]
)
```
