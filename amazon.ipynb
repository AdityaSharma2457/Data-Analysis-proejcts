```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
import time
driver = webdriver.Chrome()
driver.get("https://www.amazon.in/")
time.sleep(6)
elem = driver.find_element(By.ID, "twotabsearchtextbox")
elem.clear()
elem.send_keys("latest mobile phones 2026")
elem.send_keys(Keys.RETURN)
time.sleep(3)
html=driver.page_source
time.sleep(3)
driver.close()
with open("page.text","w",encoding="UTF-8") as f:
    f.write(html)
    
```

```python
with open("page.text","r",encoding="UTF-8") as f:
    file=f.read()
from bs4 import BeautifulSoup
soup=BeautifulSoup(file,"lxml")

elem1=soup.select("div.puisg-col-inner a.a-link-normal.s-line-clamp-2.s-line-clamp-3-for-col-12.s-link-style.a-text-normal h2.a-spacing-none span")
mobile_names=[]
for  e in elem1:
    mobile_names.append(e.text)
```

```python
elem2=soup.select("a.a-link-normal.s-line-clamp-2.s-line-clamp-3-for-col-12.s-link-style.a-text-normal")
links=[]
for e in elem2:
    links.append("amazon.in"+e.get("href"))
```

```python
elem3=soup.select("div.puisg-col-inner div.a-row span.a-price span.a-price-whole")
elem3.pop()
prices=[]
for e in elem3:
    prices.append(e.text)
```

```python
import pandas as pd

df=pd.DataFrame(columns=["product_name","price","link"])
```

```python
df['product_name']=mobile_names
df['link']=links
df['price']=prices

df=df.reset_index(drop=True)
df
```

```python
with open("count.txt","r") as f:
    c=f.read()
    f.close()
df.to_csv(f"data{c}.csv",index=False)
with open("count.txt","w") as f:
    f.write(str(int(c)+1))
    f.close()
```

```python
df2=pd.read_csv(f"data{int(c)-1}.csv") 
d1price=df["price"].tolist()
d2price=df2["price"].tolist()
d1prod=df["product_name"].tolist()
d2prod=df2["product_name"].tolist()
```

```python
for i in range(len(d1prod)):
    for j in range(len(d2prod)):
        if d1prod[i] == d2prod[j]:
            if d2price[j]< d1price[i]:
                print("price dropped !!", d1prod[i],f"from {d1price[i]} to {d2price[j]} ,\n")
```
